```
#importar pandas para ler o dataframe

import pandas as pd


NA = pd.read_csv('votacao_candidato_munzona_2018_BR.csv', encoding='latin 1', sep=';')
RN = pd.read_csv('votacao_candidato_munzona_2018_RN.csv', encoding='latin 1', sep=';')

```

```
#retirando valores nulos e organizando as colunas

NOMES_ATUAIS = ['ANO_ELEICAO', 'DT_ELEICAO', 'SG_UF', 'NM_UE', 'NM_MUNICIPIO', 'NR_ZONA', 'DS_CARGO', 'NR_CANDIDATO', 'NM_CANDIDATO', 'DS_SITUACAO_CANDIDATURA', 'DS_DETALHE_SITUACAO_CAND', 'NR_PARTIDO', 'SG_PARTIDO', 'NM_PARTIDO', 'DS_COMPOSICAO_COLIGACAO', 'DS_SIT_TOT_TURNO', 'ST_VOTO_EM_TRANSITO', 'QT_VOTOS_NOMINAIS']
NOVOS_NOMES = ['ANO', 'DATA', 'ESTADO', 'PAÍS', 'MUNICÍPIO', 'ZONA_ELEITORAL', 'CARGO', 'NUMERO_CANDIDATO', 'NOME_CANDIDATO', 'CANDIDATO', 'CANDIDATURA_APROVAÇÃO', 'PARTIDO', 'SIGLA_PARTIDO', 'NOME_PARTIDO', 'PARTIDOS_UNIDOS', 'TURNO', 'VOTO_EM_TRASIÇÃO', 'VOTOS_NOMINAIS']

NA = NA[NOMES_ATUAIS]
RN = RN[NOMES_ATUAIS]

NA.columns = NOVOS_NOMES
RN.columns = NOVOS_NOMES

NA = NA.dropna()
RN = RN.dropna()

NA


```

```
#Porcentagem dos votos dos presidentes

NA.pivot_table(index =['NUMERO_CANDIDATO'], values = ['VOTOS_NOMINAIS'], aggfunc = sum)
NA['PORCENTAGEM_PRESIDENTE'] = NA['VOTOS_NOMINAIS']*100 / NA['VOTOS_NOMINAIS'].sum()

NA
```

```
#votos do nordeste em relação aos presidentes

RIO_GRANDE = NA[NA['ESTADO'] == 'RN']
RIO_GRANDE.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

PARAIBA = NA[NA['ESTADO'] == 'PB']
PARAIBA.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

MARANHAO = NA[NA['ESTADO'] == 'MA']
MARANHAO.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

ALAGOAS = NA[NA['ESTADO'] == 'AL']
ALAGOAS.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

BAHIA = NA[NA['ESTADO'] == 'BA']
BAHIA.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

CEARA = NA[NA['ESTADO'] == 'CE']
CEARA.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

PERNAMBUCO = NA[NA['ESTADO'] == 'PE']
PERNAMBUCO.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

PIAUI = NA[NA['ESTADO'] == 'PI']
PIAUI.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

SERGIPE = NA[NA['ESTADO'] == 'SE']
SERGIPE.pivot_table(index=['ESTADO','NOME_CANDIDATO'], values = 'PORCENTAGEM_PRESIDENTE')

```

```
#Importando modulos e criando as tabelas nordeste

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))

# Load the example car crash dataset
crashes = sns.load_dataset("RIO_GRANDE").sort_values("total", ascending=False)

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="total", y="abbrev", data=crashes,
            label="Total", color="b")

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="NOME_CANDIDATO", y="PORCENTAGEM_PRESIIDENTE", data=crashes,
            label="TOP", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 24), ylabel="",
       xlabel="VOTAÇÃO PARA PRESIDENTE EM SERGIPE")
sns.despine(left=True, bottom=True)

```
